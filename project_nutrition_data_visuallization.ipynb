```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cereal_df = pd.read_csv('cereal.csv')
cereal_df.head()
#mfr: nha san xuat
#sodium: natri
#fiber: chat xo
#potass: kali
```

```python
#Irrelevant_field
fields = ['shelf', 'weight', 'cups', 'rating', 'name', 'mfr', 'type']
cereal_df_new = cereal_df.drop(fields, axis= 1);
#axis = 1: drop cols; axis = 0: drop rows
cereal_df_new.head()
```

```python
cereal_df_new.info()
```

```python
#corr()
cereal_corr = cereal_df_new.corr()#get correlation data
cereal_corr
```

```python
sns.heatmap(data=cereal_corr, cmap='Blues', linewidths=1, linecolor='Orange');
```

```python
#ones_like can build a matrix of booleans (True, False) with the same as our data
ones_corr = np.ones_like(cereal_corr, dtype = bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
#np's triu: return only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask);
#mask = mask, only take one side of triangle matrix
```

```python
adjusted_mask = mask[1:,:-1]
adjusted_mask
#comma in the center means include rows and columns
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:,:-1]
```

```python
fig, ax = plt.subplots(figsize=(10,8))

cmap = sns.diverging_palette(0,230,90,60,as_cmap=True)
#This method using HUSL colors, so you need, hue, saturation, and lightness
#I used hsluv.org to select the colors of this chart

sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask, annot=True, 
            fmt=".2f", cmap="Blues", vmin=-1, vmax=1, linecolor='white', linewidths=1);
#annot = Annotation

yticks = [i.upper()for i in adjusted_cereal_corr.index]
xticks = [i.upper()for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation = 0);
ax.set_xticklabels(xticks, rotation = 0);

title = "Correlation Matrix \n Sampled Cereals Composition \n"
ax.set_title(title);
```
